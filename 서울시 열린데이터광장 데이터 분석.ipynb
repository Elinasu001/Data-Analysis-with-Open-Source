### 1-2) API 호출, 2015년 1월부터 2024년 12월까지의 개인 유형의 형년 전기, 가스, 수도, 지역난방 에너지 사용량 데이터를 수집하는 프로그램 작성

```python
import requests
import json

api_key = "52726d4a68656c6937327048787854"
base_url = f"http://openapi.seoul.go.kr:8088/{api_key}/json/energyUseDataSummaryInfo/1/5"

# 데이터를 담을 리스트
data_list = []

for year in range(2015, 2025):  # 2015~2024
    for month in range(1, 13):  # 1~12월
        month_str = f"{month:02d}"
        url = f"{base_url}/{year}/{month_str}"

        try:
            # 데이터 수집
            response = requests.get(url)
            response.raise_for_status()
            data = response.json()

            # 정상 응답 여부 확인
            if "RESULT" in data and data["RESULT"]["CODE"] != "INFO-000":
                print(f"{year}년 {month}월 오류: {data['RESULT']['MESSAGE']}")
                continue

            # row 데이터 존재 시 처리
            rows = data.get("energyUseDataSummaryInfo", {}).get("row", [])
            for row in rows:
                if row.get("MM_TYPE") == "개인":  # 개인 유형만 출력
                    print(
                        f"{year}년 {month}월 | "
                        f"전기: {row.get('EUS', 'N/A')} | "
                        f"가스: {row.get('GUS', 'N/A')} | "
                        f"수도: {row.get('WUS', 'N/A')} | "
                        f"지역난방: {row.get('HUS', 'N/A')}"
                    )

        except requests.exceptions.RequestException as e:
            print(f"{year}년 {month}월 API 호출 실패: {e}")
        except json.JSONDecodeError as e:
            print(f"{year}년 {month}월 JSON 파싱 실패: {e}")
```

## 문제 2. 데이터 변환 및 전처리


### 2-1) 수집한 json 현태의 데이터를 pandas DataFrame 변환, 데이터 기본 정보 출력

```python
import requests
import json
import pandas as pd   # pandas 추가

api_key = "52726d4a68656c6937327048787854"
base_url = f"http://openapi.seoul.go.kr:8088/{api_key}/json/energyUseDataSummaryInfo/1/5"

# 데이터를 담을 리스트
data_list = []

for year in range(2015, 2025):  # 2015~2024
    for month in range(1, 13):  # 1~12월
        month_str = f"{month:02d}"
        url = f"{base_url}/{year}/{month_str}"

        try:
            # 데이터 수집
            response = requests.get(url)
            response.raise_for_status()
            data = response.json()

            # 정상 응답 여부 확인
            if "RESULT" in data and data["RESULT"]["CODE"] != "INFO-000":
                print(f"{year}년 {month}월 오류: {data['RESULT']['MESSAGE']}")
                continue

            # row 데이터 존재 시 추가
            rows = data.get("energyUseDataSummaryInfo", {}).get("row", [])
            for row in rows:
                if row.get("MM_TYPE") == "개인":  # 개인 유형만 수집
                    data_list.append({
                        "YEAR": row.get("YEAR"),
                        "MONTH": row.get("MON"),
                        "전기(EUS)": row.get("EUS", "N/A"),
                        "가스(GUS)": row.get("GUS", "N/A"),
                        "수도(WUS)": row.get("WUS", "N/A"),
                        "지역난방(HUS)": row.get("HUS", "N/A")
                    })

        except requests.exceptions.RequestException as e:
            print(f"{year}년 {month}월 API 호출 실패: {e}")
        except json.JSONDecodeError as e:
            print(f"{year}년 {month}월 JSON 파싱 실패: {e}")

# pandas DataFrame으로 변환
df = pd.DataFrame(data_list)

# 기본 정보 출력
print("\n===== DataFrame 기본 정보 =====")
print(df.info())
```

### 2-2) 연도별, 계절별 분석을 위해 날짜 컬럼 활용하여 연도(year)와 계절(season) 컬럼을 추가하는 전처리 코드를 작성 * 계절 구분 : 봄(3-5월), 여름(6-8월), 가을(9-11월), 겨울(12-2월)

```python
import requests
import json
import pandas as pd

api_key = "52726d4a68656c6937327048787854"
base_url = f"http://openapi.seoul.go.kr:8088/{api_key}/json/energyUseDataSummaryInfo/1/5"

data_list = []

for year in range(2015, 2025):  # 2015~2024
    for month in range(1, 13):  # 1~12월
        month_str = f"{month:02d}"
        url = f"{base_url}/{year}/{month_str}"

        try:
            response = requests.get(url)
            response.raise_for_status()
            data = response.json()

            if "RESULT" in data and data["RESULT"]["CODE"] != "INFO-000":
                print(f"{year}년 {month}월 오류: {data['RESULT']['MESSAGE']}")
                continue

            rows = data.get("energyUseDataSummaryInfo", {}).get("row", [])
            for row in rows:
                if row.get("MM_TYPE") == "개인":
                    data_list.append({
                        "YEAR": row.get("YEAR"),
                        "MONTH": row.get("MON"),
                        "전기(EUS)": row.get("EUS", "N/A"),
                        "가스(GUS)": row.get("GUS", "N/A"),
                        "수도(WUS)": row.get("WUS", "N/A"),
                        "지역난방(HUS)": row.get("HUS", "N/A")
                    })

        except requests.exceptions.RequestException as e:
            print(f"{year}년 {month}월 API 호출 실패: {e}")
        except json.JSONDecodeError as e:
            print(f"{year}년 {month}월 JSON 파싱 실패: {e}")

# pandas DataFrame으로 변환
df = pd.DataFrame(data_list)

# 날짜 기반 컬럼 추가
df["DATE"] = df["YEAR"].astype(str) + "-" + df["MONTH"].astype(str).str.zfill(2)
df["DATE"] = pd.to_datetime(df["DATE"], format="%Y-%m")

# 연도(year)와 월(month) 컬럼 추가
df["year"] = df["DATE"].dt.year

# 계절(season) 구분 함수 정의
def get_season(month):
    if month in [3, 4, 5]:
        return "봄"
    elif month in [6, 7, 8]:
        return "여름"
    elif month in [9, 10, 11]:
        return "가을"
    else:
        return "겨울"

# 계절(season) 컬럼 추가
df["season"] = df["DATE"].dt.month.apply(get_season)

# 변환 결과 확인
print("\n===== 연도 및 계절 컬럼 추가 결과 =====")
print(df[["YEAR", "MONTH", "year", "season"]].to_string(index=False))
```
